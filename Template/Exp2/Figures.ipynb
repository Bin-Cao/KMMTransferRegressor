```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df = pd.read_csv('Sourcetrain.csv')
x = df.iloc[:, 1]
```

```python
df2 = pd.read_csv('Targettrain.csv')
x2 = df2.iloc[:, 1]
```

```python
plt.hist(x, bins=6, density=True, alpha=0.6)
mu, sigma = norm.fit(x)
xmin, xmax = plt.xlim()
x_range = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x_range, mu, sigma)
plt.plot(x_range, pdf, 'b', linewidth=2)
plt.axvline(mu, color='b', linestyle='dashed', linewidth=2)
print(mu)
plt.hist(x2, bins=4, density=True, alpha=0.6)
mu2, sigma2 = norm.fit(x2)
xmin2, xmax2 = plt.xlim()
x_range2 = np.linspace(xmin2, xmax2, 100)
pdf2 = norm.pdf(x_range2, mu2, sigma2)
plt.plot(x_range2, pdf2, 'orange', linewidth=2)
plt.axvline(mu2, color='orange', linestyle='dashed', linewidth=2)
print(mu2)

plt.xlabel('x')
plt.ylabel('P(x)')
plt.savefig('Bin_distribution.png',bbox_inches = 'tight',dpi=600)
plt.savefig('Bin_distribution.svg',bbox_inches = 'tight',dpi=600)
```
